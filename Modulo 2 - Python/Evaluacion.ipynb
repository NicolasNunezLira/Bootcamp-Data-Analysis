### Ejercicio 1

Crear un sistema en el que un usuario pueda ingresar una cifra
especificando uno de los tres distintos sistemas numéricos:
- Decimal
- Binario
- Hexadecimal

El programa debe darle al usuario un menú que indique qué tipo de
número quiere ingresar, luego pedirle el número y luego debe indicar
que tipo de salida requiere. Esto se debe hacer permanentemente
hasta que el usuario ingrese la opción salir.

Debe construir una función para cada transformación, **sin** usar las
funciones de transformación numérica de Python (hex, bin,int).

```python
def decimal_to_binario(d):
    """
    Función que dado numero decimal retorna su valor en el sistema binario.
    """
    d = int(d)
    if d == 0:
        return 0
    b = ""
    while d>0:
        d, res = divmod(d,2)
        b = str(res) + b
    return b

def binario_to_decimal(b):
    """
    Función que dado numero binario en formato string retorna su valor en el sistema decimal.
    """
    d = 0
    for i in range(len(b)):
        d += int(b[i]) * (2**(len(b)-i-1))
    return d

def decimal_to_hexa(d):
    """
    Función que dado numero decimal retorna su valor en el sistema hexadecimal.
    """
    d = int(d)
    cambio = {10:"a", 11:"b", 12:"c",13:"d", 14:"e", 15:"f"}
    h = ""
    while d>0:
        d, res = divmod(d,16)
        if res in cambio:
            h = cambio[res] + h
        else:
            h = str(res) + h
    return h

def hexa_to_decimal(h):
    """
    Función que dado numero hexadecimal en formato string retorna su valor en el sistema decimal.
    """
    h = h.lower()
    cambio = {"a":10, "b":11, "c":12, "d":13, "e":14, "f":15}
    d = 0
    for i in range(len(h)):
        if h[i] in cambio:
            d += cambio[h[i]] * 16**(len(h)-i-1)
        else:
            d += int(h[i]) * 16**(len(h)-i-1)
    return d

def binario_to_hexa(b):
    """
    Función que dado numero hexadecimal en formato string retorna su valor en el sistema binario como string.
    """
    return decimal_to_hexa(binario_to_decimal(b))

def hexa_to_binario(h):
    """
    Función que dado numero binario en formato string retorna su valor en el sistema hexadecimal como string.
    """
    return decimal_to_binario(hexa_to_decimal(h))
```

```python
tipos = {"b": "binario", "d": "decimal", "h": "hexadecimal"}

while True:
    # Pregunta para nueva conversión
    yn = input("¿Desea ingresar un nuevo producto?(Y/N):")
    if yn.lower() == "n":
        break
    elif yn.lower() == "y":
        # Ingreso primer sistema numerico
        cond1 = True
        while cond1:
            print("Ingrese el sistema numerico desde el cual desea realizar la conversión\n"+
                "\t\"d\" para decimal.\n"+
                "\t\"b\" para binario.\n"+
                "\t\"h\" para hexadecimal.")
            entrada = input()
            entrada = entrada.lower()
            if entrada in tipos:
                cond1 = False
            else:
                print("Ingreso no válido. Intentelo nuevamente.\n")
        # Ingreso del numero
        print(f"Ingrese el número a realizar la conversión desde el tipo {tipos[entrada]}.")
        numero = input()
        # Ingreso segundo sistema numerico
        cond2 = True
        while cond2:
            print("Ingrese el sistema numerico hacia el cual desea realizar la conversión\n")
            salida = input()
            salida = salida.lower()
            if salida in tipos:
                cond2 = False
            else:
                print("Ingreso no válido. Intentelo nuevamente.\n")
        # Conversión
        if entrada == salida:
            print("Los tipos ingresados son los mismos.\n")
        elif entrada == "d":
            numero = int(numero)
            if salida == "b":
                print(f"El número {numero} en sistema {tipos[salida]} es {decimal_to_binario(numero)}.\n")
            elif salida == "h":
                print(f"El número {numero} en sistema {tipos[salida]} es {decimal_to_hexa(numero)}.\n")
        elif entrada == "b":
            if salida == "d":
                print(f"El número {numero} en sistema {tipos[salida]} es {binario_to_decimal(numero)}.\n")
            elif salida == "h":
                print(f"El número {numero} en sistema {tipos[salida]} es {binario_to_hexa(numero)}.\n")
        elif entrada == "h":
            if salida == "d":
                print(f"El número {numero} en sistema {tipos[salida]} es {hexa_to_decimal(numero)}.\n")
            elif salida == "b":
                print(f"El número {numero} en sistema {tipos[salida]} es {hexa_to_binario(numero)}.\n")
```

### Ejercicio 2
Crear un sistema que almacene una lista de compras. Para cada ítem,
debe almacenar el código (ej. de barras). El nombre del producto, el
precio unitario y la cantidad comprada.
Al finalizar el ingreso, se deberá emitir un informe con:

**Nombre del producto, cantidad, subtotal**

**IVA y total de la compra.**

```python
class ListaCompras():

    # Inicialización de la clase como una lista vacia
    def __init__(self):
        self.items = []

    # Agregar item
    def agregar_item(self, codigo, nombre, precio, cantidad):
        self.items.append({"Código": codigo,
                     "Nombre": nombre,
                     "Precio unitario": float(precio),
                     "Cantidad comprada": int(cantidad)})
    
    # Calculo subtotal de la lista (antes del IVA)
    def subtotal(self):
        subtotal = 0
        for i in range(len(self.items)):
            subtotal += self.items[i]["Precio unitario"] * self.items[i]["Cantidad comprada"]      
        return subtotal
    
    # Resumen de la lista
    def resumen(self):
        IVA = .19 # Variable para el IVA
        print("Resumen de la lista de compras:\n\n")
        for i, item in enumerate(self.items):
            print(f"{item['Nombre']}  x  {item['Cantidad comprada']}  =  {item['Cantidad comprada']*item['Precio unitario']}")
        print("\n --------------------------------------------------------------------------\n")
        print(f"IVA = {self.subtotal()*IVA}        Total: {self.subtotal()*(1+IVA)}")
```

```python
compras = ListaCompras()
```

```python
cond = True

while cond:
    yn = input("¿Desea ingresar un nuevo producto?(Y/N):")
    if yn.lower() == "n":
        cond = False
        compras.resumen()
    elif yn.lower() == "y":
        codigo = input("Ingrese el código del nuevo producto:")
        nombre = input("Ingrese el nombre del nuevo producto:")
        precio = input("Ingrese el precio unitario del nuevo producto:")
        cantidad = input("Ingrese la cantidad comprada del nuevo producto:")
        compras.agregar_item(codigo, nombre, precio, cantidad)
```
