### Ejercicio 1
Una compañía desea hacer predicciones del valor anual de sus ventas totales
en cierto país a partir de la relación de éstas y la renta nacional. Para
investigar la relación cuenta con los siguientes datos:

$$\begin{array}{|c|c|}
\hline
X & Y\\
\hline
189 & 402\\
190 & 404\\
208 & 412\\
227 & 425\\
239 & 429\\
252 & 436\\
257 & 440\\
274 & 447\\
293 & 458\\
308 & 469\\
316 & 469\\
\hline
\end{array}
$$

X representa la renta nacional en millones de euros e Y representa las
ventas de la compañía en miles de euros en el periodo que va desde 2010
hasta 2020 (ambos inclusive). 

1. La recta de regresión lineal de Y sobre X.

```python
# Librerias requeridas
from scipy import stats
import matplotlib.pyplot as plt

# Datos
x = [189,190,208,227,239,252,257,274,293,308,316]
y = [402,404,412,425,429,436,440,447,458,469,469]

# Regresion lineal simple
slope, intercept, r, p, std_err = stats.linregress(x,y)

# Definir la funcion de la recta
def func(x):
    return slope*x+intercept

# valores ejustados
model = list(map(func,x))

# Gráficos
plt.scatter(x,y,label='Datos')
plt.plot(x, model, label='Modelo ajustado')
plt.xlabel('Renta nacional [millones de euros]')
plt.ylabel('Ventas compañia [miles de euros]')
plt.legend()
plt.show()
```

2. El coeficiente de correlación lineal e interpretarlo.

```python
print(f'El coeficiente de relación lineal es {r:.2f}.')
```

Dado que que el coeficiente de relacion lineal es aproximandamente 1, se infiere que los datos se relacionan de manera lineal casi perfecta, y además con pendiente positiva.

3. Si en 2011 la renta nacional del país fue de 325 millones de
euros. ¿Cuál será la predicción para las ventas de la compañía
en este año?

```python
# Usamos el modelo ajustado para realizar el calculo
print(f'Bajo este escenario se predice que las ventas de ese año son {list(map(func,[325]))[0]:.2f} miles de euros.')
```

Sabiendo que el coeficiente de relación de Pearson se calcula también como:

$$ r = \dfrac{\sigma_{xy}}{\sigma_x \sigma_y}$$

1. Entonces, En una empresa de transporte trabajan 4 conductores. Los
años de antigüedad de sus permisos de conducir y el número de
infracciones cometidas en el último año por cada uno de ellos son
los siguientes:

$$\begin{array}{|c|cccc|}
\hline
X (\text{años de antigüedad}) & 3 & 4 & 5 & 6\\
\hline
Y (\text{infracciones}) & 4 & 3 & 2 & 1\\
\hline
\end{array}
$$

- Representar gráficamente los datos anteriores. Razonar si los datos
muestran una correlación positiva o negativa.

```python
# Librerias requeridas
import matplotlib.pyplot as plt

# Datos
x = range(3,7)
y = range(4,0,-1)

# Representación gráfica de los datos
plt.scatter(x,y,label='Datos')
plt.xlabel('Años de antiguedad')
plt.ylabel('Infrecciones')
plt.show()
```

Según la gráfica, se infiere que los datos se correlacionan linealmente pero con pendiente negativa, es decir, el coeficiente de correlación debe estar cercano a -1.

- Calcular el coeficiente de correlación e interpretarlo en términos de
la situación real.

```python
# Regresion lineal simple
slope, intercept, r, p, std_err = stats.linregress(x,y)

print(f'El coeficiente de correlación es {r:.2f}.')
```

Como se predijo, el coeficiente correlacion indica que los datos se relacionan de manera decreciente, es decir, a más años de antiguedad menos infracciones cometidas.

2. En un Centro de Salud hacen el seguimiento de la presión arterial a
sus pacientes, y los resultados constatan que aquéllos que tienen una
tensión arterial tienen sobrepeso, superior a la media. Calcular la covarianza y el coeficiente de correlación de Pearson de
este estudio. 

$$
\begin{array}{|c|cccccccccc|}
\hline
X \text{(Peso [kg])} & 72 & 76 & 78 & 81 & 89 & 95 & 108 & 115 & 120 & 130\\
\hline
Y \text{(Presión [mmHg])} & 115 & 121 & 125 & 130 & 141 & 150 & 165 & 170 & 177 & 178\\
\hline
\end{array}
$$

```python
# librerias requeridas
import numpy as np

# Datos
x = [72, 76, 78, 81, 89, 95, 108, 115, 120, 130]
y = [115, 121, 125, 130, 141, 150, 165, 170, 177, 178]

# Calculo de la covarianza y coeficiente correlacion
cov = np.cov(x,y)
corr = cov[0][1]/np.sqrt((cov[0][0]*cov[1][1]))

# Despliegue informacion
print(f'El coeficiente de correlación de Pearson es {corr} y la matriz de covarianza es')
print(cov)
```
