### Ejercicio 1

Continuaremos trabajando con nuestros datos de supermercado. Para
esto cargaremos los 4 archivos (clientes, boletas, detalle y productos)
y responderemos las siguientes preguntas:

1. Generar un dataframe con cada archivo.

```python
# Importamos librerias requeridas
from pandas import read_csv

# Lectura de archivos
clientes = read_csv(path+'clientes.csv', encoding='latin-1').drop_duplicates()
productos = read_csv(path+'productos.csv', encoding='latin-1').drop_duplicates()
boletas = read_csv(path+'boletas.csv', encoding='latin-1').drop_duplicates()
detalle = read_csv(path+'detalle.csv', encoding='latin-1').drop_duplicates()
```

2. ¿Cuántos datos tiene cada dataframe?

```python
print(f"El DataFrame clientes posee {len(clientes)} datos no duplicados.")
print(f"El DataFrame productos posee {len(productos)} datos no duplicados.")
print(f"El DataFrame boletas posee {len(boletas)} datos no duplicados.")
print(f"El DataFrame detalle posee {len(detalle)} datos no duplicados.")
```

3. ¿Cuántos clientes no tienen email?

```python
print(f"La cantidad de clientes que no tienen email son {clientes['email'].isna().sum()}")
```

4. Genera un filtro para saber a qué clientes les falta un dato
(cualquiera).

```python
incompletos = clientes[clientes.isna().any(axis=1)]
incompletos
```

5. ¿A cuántos clientes les faltan datos?


```python
print(f"A {len(incompletos)} clientes le falta al menos un dato.")
```

6. Agregar una nueva columna a todos los clientes, que indique si
le falta algún dato o no (con un mapeo).

```python
# Aplicamos mapeo
clientes['Completo'] = clientes.isna().any(axis=1).map({True: 'No', False: 'Si'})

# Corroboramos visualmente
clientes.head(2)
```

7. A todos los clientes que les falten datos, deben rellenarse con
un S/I, excepto la fecha de nacimiento.


```python
# Aplicamos cambios de datos faltantes NaN --> S/I
clientes[['direccion', 'comuna', 'telefono', 'email']] = clientes[['direccion', 'comuna', 'telefono', 'email']].fillna('S/I')

# Corroboramos visualmente
clientes.head(8)
```

8. Busque en los productos si existe algún outlier.

```python
# Calculamos el rango intercuartil (IQR)
Q1 = productos['precio'].quantile(0.25)
Q3 = productos['precio'].quantile(0.75)
IQR = Q3-Q1

# Calcular limites
banda_inf = Q1 - 1.5 * IQR
banda_sup = Q3 + 1.5 * IQR

# Buscamos outliers y los desplegamos
outliers = productos[(productos["precio"]>banda_sup) | (productos["precio"]<banda_inf)]

print(f"El(los) outlier(s) de los productos por precio es(son):")
for i, item in enumerate(outliers["nombre"]):
    print(f"{i+1}. {item}.")
```

6. Mostrar por pantalla el porcentaje de personas que
sobrevivieron y murieron.

```python
# Calculo de los porcentajes
porcentajes = pasajeros["Survived"].value_counts(normalize=True)

# Despliegue información
print(f"El {porcentajes[1]*100:.2f}% de los pasajeros sobrevivio al accidente, mientras que los muertos abarcan el {porcentajes[0]*100:.2f}% del total.")
```

7. Mostrar por pantalla el porcentaje de personas que sobrevivieron en cada clase.

```python
# Filtro para clases
clases = pasajeros.sort_values("Pclass")["Pclass"].unique()

for clase in clases:
    porc = pasajeros[pasajeros["Pclass"]==clase]["Survived"].value_counts(normalize=True)*100
    print(f"Sobrevivieron el {porc[1]:.2f}% de los pasajeros de clase {clase}.")
```

8. Eliminar del DataFrame los pasajeros con edad desconocida.

```python
# Eliminación pasajeros con edad desconocida
pasajeros.dropna(subset=["Age"], inplace= True)
```

9. Mostrar por pantalla la edad media de las mujeres que viajaban en cada clase.

```python
# Filtro por la variable Sex
female = pasajeros[(pasajeros['Sex']=='female')]

# Uso del filtro del item 7 para calcular las edades promedio
for clase in clases:
    print(f"La edad promedio de las mujeres de la clase {clase} es {female[female['Pclass']==clase]['Age'].mean():.2f}.")
```

10. Añadir una nueva columna booleana para ver si el pasajero era menor de edad o no.

```python
pasajeros["Adult"] = (pasajeros["Age"]>=18).map({True: 1, False:0})
pasajeros
```

11. Mostrar por pantalla el porcentaje de menores y mayores de
edad que sobrevivieron en cada clase.

```python
# Filtramos por mayores y memores de edad
mayor = pasajeros[pasajeros["Adult"]==1]
menor = pasajeros[pasajeros["Adult"]==0]

# Usamos el filtro de clase creado en 7 
for clase in clases:
    pM = mayor[mayor["Pclass"]==clase]["Survived"].value_counts(normalize=True)*100
    pm = menor[menor["Pclass"]==clase]["Survived"].value_counts(normalize=True)*100
    print(f"En la clase {clase} sobrevivieron el {pM[1]:.2f}% y {pm[1]:.2f}% de los mayores y menores de edad, resp.")
```
